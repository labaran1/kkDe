```python
df['zero'] = 0
```

```python
from sdv.constraints import GreaterThan
```

```python
cu_gt_zero = GreaterThan(
    low='zero',
    high='CU'
)
```

```python
gc = GaussianCopula(constraints=[cu_gt_zero])
```

```python
gc.fit(df)

sampled = gc.sample(2000)
```
